This method using the results of the previous tunning, then merge it together with a specific coefficient to get the new result.

```python
# Import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Input file directory
path_to_csv1 = r'<replace your path here>'
path_to_csv2 = r'<replace your path here>'
```

```python
sub1 = pd.read_csv(path_to_csv1)
sub1
```

```python
sub2 = pd.read_csv(path_to_csv2)
sub2
```

```python
# Select only ID column and Target column
sub1 = sub1[['SK_ID_CURR', 'TARGET']]
sub2 = sub2[['SK_ID_CURR', 'TARGET']]

# Concat 2 submission files by SK_ID_CURR, to avoid the problem that diffirent ordered of ID columns in 2 files
sub1.columns = ['SK_ID_CURR','1']
sub2.columns = ['SK_ID_CURR','2']
df_subs = pd.merge(sub1,sub2,on='SK_ID_CURR',how='left')
df_subs
```

```python
# check correlation map between these 22 best AUC scores
plt.figure(figsize=(6,6))
sns.heatmap(df_subs.iloc[:,1:].corr(), cmap='Spectral')
plt.show()
```

If lowest values of correlation is high, it means that 2 datasets is very similar and the blending method may not improve scores much

```python
# create a submission dataset
calculate = pd.DataFrame()
calculate['SK_ID_CURR'] = df_subs['SK_ID_CURR']

# create a linear combination between 2 individual results 
# coefficients determine empirically and based on correlational interactions
# the best coefficients is based on trials, which means we need to test several coefficients to know which is the best
calculate['TARGET'] = 0.3615*df_subs['1'] + 0.6385*df_subs['2']

calculate.head()
```

```python
calculate.to_csv('blending.csv')
```
